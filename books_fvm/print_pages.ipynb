```python
def generar_cadena_pares(inicio, fin):
    """
    Genera un string de números (separados por coma y espacio) tomando un 
    valor inicial y uno final, soportando secuencias ascendentes y descendentes.
    Retorna la cadena resultante y la cantidad total de números.
    """
    # Determinamos la dirección (ascendente o descendente)
    if inicio <= fin:
        paso = 2
        limite_ajustado = fin + 1 # Para incluir el número final si coincide
    else:
        paso = -2
        limite_ajustado = fin - 1 # Para incluir el número final hacia abajo
        
    # Creamos una lista con los números para poder contarlos fácilmente
    numeros = list(range(inicio, limite_ajustado, paso))
    
    # Obtenemos la cantidad de números
    cantidad = len(numeros)
    
    # Convertimos cada número a string y los unimos con ", "
    resultado = ", ".join(str(numero) for numero in numeros)
    
    return resultado, cantidad

# Parámetros de prueba basados en tu ejemplo (descendente)
valor_inicial = 265
valor_final = 237

# Ejecutamos la función
cadena_resultante, cantidad_numeros = generar_cadena_pares(valor_inicial, valor_final)

# Mostramos los resultados
print(f'Cadena resultante: "{cadena_resultante}"')
print(f'Cantidad de números en la cadena: {cantidad_numeros}')

# === Prueba extra para comprobar el modo ascendente ===
# print("\n--- Prueba Ascendente ---")
# cad_ascendente, cant_ascendente = generar_cadena_pares(236, 250)
# print(f'Cadena: "{cad_ascendente}" | Cantidad: {cant_ascendente}')
```
